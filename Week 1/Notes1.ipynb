Sentence -> Assertion about the world in knowledge represetnation language (ex, it rained). It is structured in formal language (Propositional logic, first-order logic...) to express facts, relationships, or rules about the world in a way that a computational system can process.

### Propositional Logic 

* based on propositions, statements that are True/False
    * Propositional Symbols: Letters denoting some event, object (P,Q,R...)
    * Logical Connectives: 
                            NOT()
                            AND()
                            OR()
                            Implication()
                            BiConditional()

* Knowledge Base (KB)
    * set of sentences known to be true by the agent

* Model 
    * assignment of True/False to every proposition about the world, according to the knowledge base.
        * ex: "it is raining", "it is monday" -> P = T, Q = T
    
* Entailment 
    * Similar to implication, but implication is between 2 propositions, while entailment is between two blobs of of info. In more formal sense, it is a fundamental concept in logic, referring to a relationship between a set of premises (in this case, the knowledge base) and a conclusion.
        * ex: if KB is true, then B is true. 

### Model Checking

* we enumerate all possibilities to check whether our proposition is true, given our knowledge base. 

### Inference Rules 

| Inference Rule                | Mathematical Expression | Description |
|-------------------------------|-------------------------|-------------|
| Modus Ponens                  | &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $`  \begin{array}{c} \alpha \rightarrow \beta \\ \alpha \\ \hline \beta \end{array} `$ | If "A implies B" is true and A is true, then B must also be true. |
| And Elimination               | &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $`\begin{array}{c} \alpha \land \beta \\ \hline \alpha \end{array} `$ | If an 'And' proposition is true, any one element of it is true. |
| Double Negation Elimination   | &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $`\begin{array}{c} \neg(\neg\alpha) \\ \hline \alpha \end{array} `$ | A double negation cancels out, making the statement true. |
| Implication Elimination       | &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $`\begin{array}{c} \alpha \rightarrow \beta \\ \hline \neg \alpha \lor \beta \end{array}`$ | An implication can be rephrased as 'not A or B'. |
| Biconditional Elimination     | &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $`\begin{array}{c} \alpha \leftrightarrow \beta \\ \hline (\alpha \rightarrow \beta) \land (\beta \rightarrow \alpha) \end{array} `$ | A biconditional splits into two implications. |
| De Morgan’s Law               | &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $`\begin{array}{c} \neg(\alpha \land \beta) \\ \hline \neg \alpha \lor \neg \beta \end{array} `$ | The negation of 'And' is 'Or' with negated components. |
|                               | &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $`\begin{array}{c} \neg(\alpha \lor \beta) \\ \hline \neg \alpha \land \neg \beta \end{array} `$ | The negation of 'Or' is 'And' with negated components. |
| Distributive Property         | &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $`\begin{array}{c} \alpha \land (\beta \lor \gamma) \\ \hline (\alpha \land \beta) \lor (\alpha \land \gamma) \end{array} `$ | 'And' distributes over 'Or'. |
|                               | &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $`\begin{array}{c} \alpha \lor (\beta \land \gamma) \\ \hline (\alpha \lor \beta) \land (\alpha \lor \gamma) \end{array} `$ | 'Or' distributes over 'And'. |
| Resolution                    | &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $`\begin{array}{c} P \lor Q \\ \neg P \\ \hline Q \end{array} `$ | If one of two propositions in an Or proposition is false, the other must be true. |
|                               | &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $`\begin{array}{c} P \lor Q \\ \neg P \lor R \\ \hline Q \lor R \end{array} `$ | Resolution can be applied to deduce a new Or proposition when one part of the initial Or proposition is negated. |

In a way, knowledge can also be seen as a search problem where the knowledge problem maps onto search problems as: 

- Initial State: Starting KB
- Actions: Inference Rules
- Transition Model: new KB after inference
- Goal Test: Checking statement that we are trying to prove in the KB
- path cost: number of steps to proof

### Resolution Refutation Overview

Resolution Refutation is a potent method used in logic and artificial intelligence to determine if a knowledge base (KB) entails a specific conclusion, denoted as 'X'. The process involves transforming the KB into its Conjunctive Normal Form (CNF) and then systematically attempting to produce an empty clause. Achieving an empty clause signifies that the knowledge base indeed entails 'X'.

#### Key Concepts

- **disjunction**: propositions connected with an Or logical connectives (P ∨ Q ∨ R)
- **conjunction**: propositions connected with an And logical connectives (P ∧ Q ∧ R).
- **Clause**: A disjunction (OR) of literals, where a literal is either a propositional symbol or its negation. A clause could be a single literal (P) or a combination of several literals connected by OR (P ∨ ¬Q ∨ R).
- **Conjunctive Normal Form (CNF)**: conjunction (AND) of clauses, for example: (A ∨ B ∨ C) ∧ (D ∨ ¬E) ∧ (F ∨ G)

#### Conversion to CNF

1. **Eliminate Biconditionals**: Transform (α ↔ β) into (α → β) ∧ (β → α).
2. **Eliminate Implications**: Convert (α → β) into ¬α ∨ β.
3. **Move Negation Inwards**: Apply De Morgan’s Laws to ensure negations only apply to literals.
4. **Apply Distributive Law**: To distribute OR over AND, making the expression suitable for resolution.

#### Resolution Refutation Process

- **Factoring**: Remove duplicate literals within clauses to simplify the expressions.
- **Resolving Literals**: The process where a literal and its negation, when resolved, produce the empty clause (), indicating a contradiction.
- **Empty Clause**: The goal of resolution refutation, symbolizing that the KB cannot simultaneously be true and false, thus proving entailment.

#### Proving Entailment

To prove if KB entails 'X':

1. **Contradiction Check**: Add ¬'X' to KB, convert to CNF, and seek to produce an empty clause through resolution.
2. **Derive Empty Clause**: The derivation of the empty clause confirms that KB entails 'X'.
3. **Failure to Derive Empty Clause**: Indicates no entailment between KB and 'X'.


### First Order Logic

* Propositional logic may get too verbose, as we don't need to express each variation of an event with a different symbol. instead we can use 2 types of symbols and thro combinations of those symbols produce the needed events. 

The 2 types of symbols are: 
* Constant Symbols: Represent Objects, (Luka, KIU, etc)
* Predicate Symbols: Relations/Functions that take arguments and return True/False. (Person(Luka), University(KIU), BelongsTo(Luka, KIU))

for example, ~University(Luka) conveys that Luka is not an university. 



- Universal quantification: for all x, ....
- Existential quantification: there is at least 1 x, such that ....


To illustrate how model checking achieves **1) Consistency, 2) Entailment Verification,** and **3) Comprehensive Validation**, let's consider a simplified and arbitrary Knowledge Base (KB) concerning weather conditions:

### Knowledge Base (KB):
1. If it is raining (R), then the ground is wet (W).
2. It is not raining (¬R) or it is cloudy (C).

In logical form:
1. `R → W`
2. `¬R ∨ C`

### Goal:
We want to use model checking to verify if the KB entails that "If it is cloudy, then the ground is wet" `(C → W)`.

### Step 1: Enumerate All Possible T/F Permutations

For the symbols R, W, and C, we enumerate all possible truth values:

| R   | C   | W   | KB  |
|-----|-----|-----|-----|
| T   | T   | T   | ?   |
| T   | T   | F   | ?   |
| T   | F   | T   | ?   |
| T   | F   | F   | ?   |
| F   | T   | T   | ?   |
| F   | T   | F   | ?   |
| F   | F   | T   | ?   |
| F   | F   | F   | ?   |

### Step 2: Mark True in the KB Column Where the KB Holds

Apply the rules of the KB to each permutation to determine if the KB holds:

- `R → W` holds true when R is false, or both R and W are true.
- `¬R ∨ C` holds true unless R is true and C is false.

| R   | C   | W   | KB  | Comment                           |
|-----|-----|-----|-----|-----------------------------------|
| T   | T   | T   | T   | Both conditions are satisfied.    |
| T   | T   | F   | F   | `R → W` is violated.              |
| T   | F   | T   | F   | `¬R ∨ C` is violated.             |
| T   | F   | F   | F   | Both conditions are violated.     |
| F   | T   | T   | T   | Both conditions are satisfied.    |
| F   | T   | F   | T   | `¬R ∨ C` is satisfied.            |
| F   | F   | T   | T   | `¬R ∨ C` is satisfied.            |
| F   | F   | F   | T   | `¬R ∨ C` is satisfied.            |

### Step 3: Verify Entailment of `C → W`

Now, we check if `C → W` (If it is cloudy, then the ground is wet) is true in all models where the KB is true:

| R   | C   | W   | KB  | `C → W` | Comment                                         |
|-----|-----|-----|-----|---------|-------------------------------------------------|
| T   | T   | T   | T   | T       | `C → W` holds.                                  |
| F   | T   | T   | T   | T       | `C → W` holds.                                  |
| F   | T   | F   | T   | F       | `C → W` does not hold.                          |
| F   | F   | T   | T   | T       | Irrelevant, as C is false.                      |
| F   | F   | F   | T   | T       | Irrelevant, as C is false.                      |

### Evaluation:

1. **Consistency:** The KB itself is consistent, as there are models (truth assignments) where the KB holds true. There are no internal contradictions within the KB's rules.
   
2. **Entailment Verification:** The KB does not entail `C → W` in all models where the KB is true, as shown by the counterexample where C is true and W is false but the KB still holds. Thus, the claim "If it is cloudy, then the ground is wet" is not entailed by the KB.
   
3. **Comprehensive Validation:** Through the process of enumerating all T/F permutations and applying the KB's conditions, we've validated the logical outcomes of the KB comprehensively. This method ensured that we considered all possible scenarios, including those not directly stated but logically implied by the KB.

This example demonstrates how model checking serves as a robust tool for verifying logical consistencies, entailments, and comprehensive validation of knowledge bases in logical systems.
